In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd
url_1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url_2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url_3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df1 = pd.read_csv(url_1)
df2 = pd.read_csv(url_2)
df3 = pd.read_csv(url_3)
print(df1.columns)
print(df2.columns)
print(df3.columns)

# pd.concat([df1, df2, df3], axis=0).reset_index(drop=True)
```

```python
#Cleaning

df1 = df1.applymap(lambda x: x.lower() if isinstance(x, str) else x)
df1 = df1.dropna(axis=0, how='all')  # Remove rows with all NaN values
df1 = df1.drop_duplicates()  # Drop duplicates
df1.columns = df1.columns.str.lower()  # make column names lowercase
df1.columns = df1.columns.str.strip()  # strip leading/trailing whitespace
df1.columns = df1.columns.str.replace(' ', '_')  # replace spaces with underscores
df1 = df1.reset_index(drop=True)
#DF3
df2 = df2.applymap(lambda x: x.lower() if isinstance(x, str) else x)
df2 = df2.dropna(axis=0, how='all')  # Remove rows with all NaN values
df2 = df2.drop_duplicates()  # Drop duplicates
df2.columns = df2.columns.str.lower()  # make column names lowercase
df2.columns = df2.columns.str.strip()  # strip leading/trailing whitespace
df2.columns = df2.columns.str.replace(' ', '_')  # replace spaces with underscores
df2 = df2.reset_index(drop=True)
#DF3
df3 = df3.applymap(lambda x: x.lower() if isinstance(x, str) else x)
df3 = df3.dropna(axis=0, how='all')  # Remove rows with all NaN values
df3 = df3.drop_duplicates()  # Drop duplicates
df3.columns = df3.columns.str.lower()  # make column names lowercase
df3.columns = df3.columns.str.strip()  # strip leading/trailing whitespace
df3.columns = df3.columns.str.replace(' ', '_')  # replace spaces with underscores
df3 = df3.reset_index(drop=True)
```

```python
#Cleaning block
print(df1.columns)
print(df2.columns)
print(df3.columns)
```

```python
# Concatenate all dataframes along rows
df4 = pd.concat([df1, df2, df3], axis=0)
df4 = df4.reset_index(drop=True)
df4 = df4.fillna('')
print(df4.columns)

df4.to_csv('cleaned_combined_dataset.csv', index=False)
```

```python
df4['gender'] = df4['gender'].replace({'m': 'm', 'male': 'm', 'F': 'f', 'female': 'f', 'femal': 'f', 'female': 'f'})
# df4['number_of_open_complaints'] = df4['number_of_open_complaints'].str.split('/').str[1].astype('int')
df4 = df4.rename(columns={'st': 'state'})
df4['state'] = df4['state'].replace({'wa': 'washington', 'az': 'arizona'})
df4['education'] = df4['education'].replace({'bachelors': 'bachelor', })
df4['vehicle_class'] = df4['vehicle_class'].replace({'luxury Car': 'luxury', 'sports car': 'luxury', 'luxury suv': 'luxury'})
df4['customer_lifetime_value'] = df4['customer_lifetime_value'].str.strip('%')
# set(df5['education'])
# set(df5['gender'])
# df5.nunique()

# set(df5['sales_channel'])
# set(df5['month'])
df4.head(50)
```
